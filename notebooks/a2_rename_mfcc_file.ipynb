```python
config = load_config('configs/configs.yaml')
```

```python

# Define the folder path
folder_path = config["MFCC_FOLDER"]

# List all files in the folder
files = os.listdir(folder_path)

# Iterate through each file
for file_name in files:
    if file_name.startswith('hate_video_') and file_name.endswith('_hate_mfcc.p'):
        # Extract the number x from the file name
        parts = file_name.split('_')
        x = parts[2]
        
        # Construct the new file name
        new_file_name = f'hate_video_{x}_mfcc.p'
        
        # Rename the file
        os.rename(os.path.join(folder_path, file_name), os.path.join(folder_path, new_file_name))
        print(f'Renamed {file_name} to {new_file_name}')
    if file_name.startswith('non_hate_video_') and file_name.endswith('non_hate_mfcc.p'):
        # Extract the number x from the file name
        parts = file_name.split('_')
        x = parts[3]
        
        # Construct the new file name
        new_file_name = f'non_hate_video_{x}_mfcc.p'
        
        # Rename the file
        os.rename(os.path.join(folder_path, file_name), os.path.join(folder_path, new_file_name))
        print(f'Renamed {file_name} to {new_file_name}')
```
